The following results accompany the sensitivity analyses presented in Supplement S7 of the paper ([SSRN](https://papers.ssrn.com/sol3/papers.cfm?abstract_id=3635047), [MedRxiv](https://www.medrxiv.org/content/10.1101/2020.06.24.20139451v1)).

Specifically, here we report country-level and global changes in 12 key outcome measures (6 country-level, 6 global), as detailed in Supplement S7.

Results for tests of parametric assumptions are reported as *elasticities*, i.e. % change in outcome per % change in parameter value for +/-5% changes in assumed values. These elasticities are summarised in Table S7 of the Supplement.

Results for other analyses are reported as *sensitivities*, i.e. % change in outcome from the baseline scenario. These analyses include changes in assumed parameter variances (StdScale, reported in Tables S5 and S6 of the Supplement) and country subsamples (exclusion of top countries by population, actual cases, and reported cases, summarised in Table S8).


```python
import pandas as pd

a = 0
with open("sensOutputs.txt", 'r') as output:
    reader = output.read()
    for i, part in enumerate(reader.split("\n\n\n\n")):
        if i == 0:
            pass
        else:
            for j, subpart in enumerate(part.split("\n\n")):
                with open(f"sensOutput{i}{j}.txt", "w") as f:
                    f.write(subpart)
        a += 1

for i in range(a)[1:]:
    for j in [0,1]:
        with open(f"sensOutput{i}{j}.txt", 'r') as op:
            lines = op.readlines()
            title = lines[0][:-1]
            header = lines[1].split('\t')
            columns = list(zip([title] * len(header), header))
            columns = pd.MultiIndex.from_tuples(columns, names=None)

        with open(f"sensOUtput{i}{j}.txt", 'r') as op:
            df = pd.read_csv(op, sep='\t', header=1)
            df.columns = columns

        display(df.style.format("{:.3g}", subset=columns[1:]).\
                set_table_styles([{'selector':'th', 'props':[('text-align', 'center')]}]).\
                hide_index())
    print("\n\n")
```
